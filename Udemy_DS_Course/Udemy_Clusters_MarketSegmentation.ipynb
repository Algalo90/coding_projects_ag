```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
# Load the Data:

data = pd.read_csv('3.12. Example.csv')
data
# Each observation is a client.
# Satisfaction is self reported, from 1 to 10
# Loyalty is a derivative variable calculatides with the number of purchases per year and similar indicators (-2.5 to 2.5 (it is already standartized))
```

```python
# Plotting the data:

plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
# Select the features

x = data.copy()
```

```python
# Clustering

kmeans = KMeans(2)
kmeans.fit(x)
```

```python
# Clustering results:

clusters = x.copy()
clusters['cluster pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c=clusters['cluster pred'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
# As wee didn't standardize satisfaction, and satisfaction values are higher than those of Loyalty,
# The KMeans function has only used Satisfaction to create the clusters
```

```python
# Standartice the values:
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
# Take advantage of the elbow method
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('nº of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
# Explore clustering solutions and select the number of clusters

kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c=clusters_new['cluster_pred'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

## How is Clustering useseful:

- Exploratory Analysis:
    - Getting acquanted with data
    - Searching for patterns
    - Deciding what methods to utilize
    
- Confirmatory and explanatory analysis:
    - Explain a phenomenon
    - Confirm hypothesis
    - Validate previous research
    
Clustering can be used for all analyses but it is more frequently used in Exploratory analysis.
